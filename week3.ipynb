```python
import mysql.connector
from contextlib import closing
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
dbconfig = {
    'user': os.getenv('MYSQL_USER'),
    'password': os.getenv('MYSQL_PASSWORD'),
    'host': 'localhost',
    'port': '3306',
    'database': 'LittleLemonDB'
}
```

## Task 1 - Connecting with the database

```python
connection = mysql.connector.MySQLConnection(**dbconfig)
```

```python
cursor = connection.cursor()
```

## Task 2 - Query the database

```python
cursor.execute(
    '''
    SHOW TABLES;
    '''    
)
cursor.fetchall()
```

## Task 3 - Query with table JOIN

```python
cursor.execute(
    '''
    SELECT DISTINCT Customers.CustomerName AS FullName, Customers.ContactNumber
    FROM Customers
    INNER JOIN Orders
    USING (CustomerID)
    WHERE Orders.TotalCost > 60;
    '''
)
print('Sno.', cursor.column_names)
for i, result in enumerate(cursor):
    print(i, result)
```
